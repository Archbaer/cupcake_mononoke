```python
import os 
from dotenv import load_dotenv

load_dotenv()
alpha_vantage_key = os.getenv("ALPHA_VANTAGE")
```

```python
import requests

class QueryAlphaVantage:
    def __init__(self, api_key):
        self.api_key = api_key

    def get_data(self, symbol, function):
        base_url = "https://www.alphavantage.co/query"
        params = {
            "function": function,
            "symbol": symbol,
            "apikey": self.api_key
        }
        response = requests.get(base_url, params=params)
        if response.status_code == 200:
            return response.json()
        else:
            response.raise_for_status()
```

```python
api = QueryAlphaVantage(alpha_vantage_key)
data = api.get_data("AAPL", "TIME_SERIES_DAILY")
print(data)
```

```python
energy_tickers = ['XOM', 'CVX', 'COP', 'SLB', 'EOG', 'MPC', 'PSX', 'VLO']

# Get overview data for each
api = QueryAlphaVantage(alpha_vantage_key)
companies = []

for ticker in energy_tickers:
    overview = api.get_data(ticker, "OVERVIEW")
    if 'MarketCapitalization' in overview:
        companies.append({
            'Symbol': ticker,
            'Name': overview.get('Name'),
            'MarketCap': int(overview.get('MarketCapitalization', 0))
        })
```

```python
# Sort by market cap
import pandas as pd
df_energy = pd.DataFrame(companies).sort_values('MarketCap', ascending=False)
df_energy
```
